**LAB1: Regression**

**1.1 Simple Linear Regression**  
Using a single continuous feature (e.g., Cholesterol_Total) to predict Age.

```python
%pip install nbformat ipython plotly --upgrade
```

```python
import pandas as pd
import numpy as np
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# 1. Load Dataset
df = pd.read_csv(r'C:\ML-CPE\LAB02\dataset\synthetic_heart_disease_dataset.csv')

# 2. Setup Data & Train Model for Age Prediction
feature_cols = ['Cholesterol_Total', 'Systolic_BP', 'Diastolic_BP', 'Heart_Rate']
feature_cols = [col for col in feature_cols if col in df.columns]

X_multi = df[feature_cols].fillna(df[feature_cols].median()) # Clean missing on the fly
y_age = df['Age']

X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(X_multi, y_age, test_size=0.2, random_state=42)

multi_lr = LinearRegression()
multi_lr.fit(X_train_m, y_train_m)
y_pred_multi = multi_lr.predict(X_test_m)

# 3. Plot Result using Plotly
plot_data = pd.DataFrame({
    'Actual Age': y_test_m,
    'Predicted Age': y_pred_multi
})

fig = px.scatter(
    plot_data, 
    x='Actual Age', 
    y='Predicted Age', 
    trendline="ols",
    title="Age Prediction: Actual vs Predicted"
)

fig.show(renderer="vscode")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Define feature (X) and target variable (y) for Simple Linear Regression
X_simple = df[['Cholesterol_Total']]
y_age = df['Age']

# Split data into training and testing sets (80% Train, 20% Test)
X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(X_simple, y_age, test_size=0.2, random_state=42)

# Initialize and train Simple Linear Regression model
simple_lr = LinearRegression()
simple_lr.fit(X_train_s, y_train_s)

# Predict on test data
y_pred_simple = simple_lr.predict(X_test_s)

# Display evaluation metrics
print("--- Simple Linear Regression Results ---")
print(f"Coefficient (Slope) : {simple_lr.coef_[0]:.4f}")
print(f"Intercept           : {simple_lr.intercept_:.4f}")
print(f"Mean Squared Error  : {mean_squared_error(y_test_s, y_pred_simple):.4f}")
print(f"R-squared Score     : {r2_score(y_test_s, y_pred_simple):.4f}")
```

**1.2 Multiple Linear Regression**  
Using multiple clinical metrics (e.g., Cholesterol_Total, Systolic_BP, Diastolic_BP, Heart_Rate) to predict Age.

```python
# Define multiple features (X) for Multiple Linear Regression
feature_cols = ['Cholesterol_Total', 'Systolic_BP', 'Diastolic_BP', 'Heart_Rate']
# Filter only columns that exist in the dataframe
feature_cols = [col for col in feature_cols if col in df.columns]

X_multi = df[feature_cols]

# Split data into training and testing sets
X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(X_multi, y_age, test_size=0.2, random_state=42)

# Initialize and train Multiple Linear Regression model
multi_lr = LinearRegression()
multi_lr.fit(X_train_m, y_train_m)

# Predict on test data
y_pred_multi = multi_lr.predict(X_test_m)

# Display evaluation metrics
print("--- Multiple Linear Regression Results ---")
print("Coefficients        :")
for col, coef in zip(feature_cols, multi_lr.coef_):
    print(f"  - {col}: {coef:.4f}")
print(f"Intercept           : {multi_lr.intercept_:.4f}")
print(f"Mean Squared Error  : {mean_squared_error(y_test_m, y_pred_multi):.4f}")
print(f"R-squared Score     : {r2_score(y_test_m, y_pred_multi):.4f}")
```

**LAB 2: Classification**

**2.1 Preparing Classification Data**

Preparing features and target variable for classification.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Select target variable: 'Heart_Disease' (or 'Gender' if present in your dataset)
target_col = 'Heart_Disease' if 'Heart_Disease' in df.columns else 'Gender_Encoded'

# Select 2 continuous features for decision boundary visualization
feature_cols_clf = ['Cholesterol_Total', 'Systolic_BP']

# Extract X and y
X_clf = df[feature_cols_clf].fillna(df[feature_cols_clf].median())
y_clf = df[target_col]

# Split data into training and testing sets (80% Train, 20% Test)
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(X_clf, y_clf, test_size=0.2, random_state=42)

# Feature Scaling (Recommended for Logistic Regression)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_c)
X_test_scaled = scaler.transform(X_test_c)

print(f"Classification target selected: {target_col}")
print(f"Feature matrix shape: {X_train_scaled.shape}")
```

---

**2.3 Logistic Regression**  
Training a Logistic Regression classifier on the prepared data.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Initialize and train Logistic Regression model
clf_model = LogisticRegression(random_state=42)
clf_model.fit(X_train_scaled, y_train_c)

# Predict on test set
y_pred_c = clf_model.predict(X_test_scaled)

# Display accuracy score
accuracy = accuracy_score(y_test_c, y_pred_c)
print(f"Logistic Regression Accuracy: {accuracy * 100:.2f}%")
```

**2.4 Confusion Matrix**  
Evaluating classification performance using a Confusion Matrix.

```python
from sklearn.metrics import confusion_matrix, classification_report
import plotly.express as px

# Compute Confusion Matrix
cm = confusion_matrix(y_test_c, y_pred_c)

# Display Classification Report
print("--- Classification Report ---")
print(classification_report(y_test_c, y_pred_c))

# Plot Heatmap Confusion Matrix using Plotly
fig_cm = px.imshow(
    cm,
    text_auto=True,
    labels=dict(x="Predicted Label", y="Actual Label", color="Count"),
    x=['Class 0', 'Class 1'],
    y=['Class 0', 'Class 1'],
    title="Confusion Matrix Heatmap",
    color_continuous_scale="Blues"
)

fig_cm.show()
```

**LAB 3: Model Comparison**

**3.1 Simple vs Multiple Linear Regression**  
Comparing regression models trained on a single feature vs multiple features for Age Prediction.

```python
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# Evaluate Simple Linear Regression
mse_simple = mean_squared_error(y_test_s, y_pred_simple)
rmse_simple = np.sqrt(mse_simple)
r2_simple = r2_score(y_test_s, y_pred_simple)

# Evaluate Multiple Linear Regression
mse_multi = mean_squared_error(y_test_m, y_pred_multi)
rmse_multi = np.sqrt(mse_multi)
r2_multi = r2_score(y_test_m, y_pred_multi)

# Summary Comparison Table
reg_comparison = pd.DataFrame({
    'Model': ['Simple Linear Regression', 'Multiple Linear Regression'],
    'MSE': [mse_simple, mse_multi],
    'RMSE': [rmse_simple, rmse_multi],
    'R2 Score': [r2_simple, r2_multi]
})

print("--- Regression Model Comparison ---")
print(reg_comparison.to_string(index=False))
```
